```python
import nltk
from nltk import FreqDist
```

```python
import pandas as pd

df=pd.read_csv (r'News_Article_2016_Jan.csv')
```

```python
content_list=df['Content']

content=' '.join(content_list)
```

```python
content = content.replace('\n',' ') 
```

```python
type(content)
```

#Q1 simple bag of words

```python
tokens = nltk.word_tokenize(content)
type(tokens)
```

```python
#change all tokens into lower case 
words1 = [w.lower() for w in tokens]   #list comprehension 

#only keep text words, no numbers 
words2 = [w for w in words1 if w.isalpha()]
```

```python
#generate a frequency dictionary for all tokens 
freq = FreqDist(words2)

#sort the frequency list in descending order
sorted_freq = sorted(freq.items(),key = lambda k:k[1], reverse = True)
sorted_freq
```

```python
#If the this code does not show a figure, just try to run it twice
freq.plot(30)
```

#2.bag-of-words approach with stemming and stop words removal

```python
from nltk.corpus import stopwords

stopwords = stopwords.words('english') #use the NLTK stopwords
```

```python
#only keep the words that not in nltk stopwords word list
words_nostopwords = [w for w in words2 if w not in stopwords]
```

```python
#generate a frequency dictionary for all tokens 
freq_nostw = FreqDist(words_nostopwords)

#sort the frequency list in decending order
sorted_freq_nostw = sorted(freq_nostw.items(),key = lambda k:k[1], reverse = True)
sorted_freq_nostw
```

```python
#If the this code does not show a figure, just try to run it twice
freq_nostw.plot(30)
```

```python
#Steps are the same with the Porter Stemmer
lancaster = nltk.LancasterStemmer()
stem = [lancaster.stem(w) for w in words_nostopwords]
stem = [w for w in stem]
freq = FreqDist(stem)
sorted_freq = sorted(freq.items(),key = lambda k: k[1], reverse = True)
```

```python
sorted_freq
```

```python
freq.plot(30)
```

#3 Use POS approach and focus on all the noun forms (NN, NNP, NNS, NNPS)

```python
nltk.download('averaged_perceptron_tagger')
```

```python
POS_tags = nltk.pos_tag(tokens) #use unprocessed 'tokens', not 'words'

#Generate a list of POS tags
POS_tag_list = [(word,tag) for (word,tag) in POS_tags if tag.startswith('N')]
```

```python
#Generate a frequency distribution of all the POS tags
tag_freq = nltk.FreqDist(POS_tag_list)
#Sort the result 
sorted_tag_freq = sorted(tag_freq.items(), key = lambda k:k[1], reverse = True)
```

```python
sorted_tag_freq
```

```python
tag_freq.plot(30)
```

```python
#write result into .txt file
with open('POS_output.txt','w', encoding="utf-8") as f:
    for (word,tag),frequency in sorted_tag_freq:
        f.write(str(word)+'\t'+str(tag)+'\t'+str(frequency)+'\n')
```

```python
import warnings
warnings.filterwarnings('ignore')
import re
from nltk.stem import WordNetLemmatizer, PorterStemmer, SnowballStemmer
stop_words_file = 'SmartStoplist.txt'

stop_words = []

with open(stop_words_file, "r") as f:
    for line in f:
        stop_words.extend(line.split()) 
        
stop_words = stop_words  

def preprocess(raw_text):
    
    #regular expression keeping only letters 
    letters_only_text = re.sub("[^a-zA-Z]", " ", str(raw_text))

    # convert to lower case and split into words -> convert string into list ( 'hello world' -> ['hello', 'world'])
    words = letters_only_text.lower().split()

    cleaned_words = []
    lemmatizer = PorterStemmer() #plug in here any other stemmer or lemmatiser you want to try out
    
    # remove stopwords
    for word in words:
        if word not in stop_words:
            cleaned_words.append(word)
    
    # stemm or lemmatise words
    stemmed_words = []
    for word in cleaned_words:
        word = lemmatizer.stem(word)   #dont forget to change stem to lemmatize if you are using a lemmatizer
        stemmed_words.append(word)
    
    # converting list back to string
    return " ".join(stemmed_words)
```
